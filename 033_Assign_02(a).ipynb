```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# Find out how many principal components explain more than 90% variation taking all the variables except ID and target. 
```

```python
df = pd.read_csv("train_Santander.csv")
df
```

```python
testdf = pd.read_csv("test_Santander.csv",index_col = 0)
testdf
```

```python
df["TARGET"].unique()
```

```python
# define X and y

X = df.drop(["TARGET"],axis = 1)
y = df["TARGET"]
```

```python

X_train, y_train, X_test, y_test = train_test_split(X,y,random_state = 24, test_size=0.3)
```

```python
lr = LogisticRegression()
```

```python
scaler  = StandardScaler().set_output(transform='pandas')
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
pca = PCA().set_output(transform = "pandas")
p_comps = pca.fit_transform(X_scaled)

pc_cumsum = np.cumsum(pca.explained_variance_ratio_ * 100)
pc_cumsum
```

```python
#a. Finding out how many principal components explain more than 90% variation  
print((len(pc_cumsum)- (pc_cumsum > 90).sum()) + 1 )
```

```python
y_pred = pca.predict(X_test)
```

```python
res = pd.DataFrame({"id": list(testdf.index), "target":y_pred })
res.to_csv("Santander_pca.csv", index = False)
```
