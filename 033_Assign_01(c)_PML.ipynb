```python
import numpy as np
import pandas as pd

from sklearn.model_selection import KFold,GridSearchCV
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv("abalone_train.csv",index_col = 0)
train
```

```python
testdf = pd.read_csv("abalone_test.csv",index_col = 0)
```

```python
######################## XGBoost ###############################
```

```python
# as our y is multi-class col == use le;

le =LabelEncoder()
y = le.fit_transform(train["Rings"])
le.classes_
```

```python
# define X:
# as we can see "sex" col == > catergorical data ==> convert into numerical

X = train.drop(["Rings"],axis = 1)
# crete dumies:
dum_X = pd.get_dummies(X,drop_first = True)   
dum_X 
```

```python
# create obj
xgb = XGBRegressor(random_state = 24)
```

```python
kfold = KFold(n_splits = 5, shuffle = True, random_state = 24)
params = {"max_depth":[None,3,4], "learing_score": np.linspace(0.001,0.9,10),'n_estimators' : [25, 50, 100]}
gcv = GridSearchCV(xgb,param_grid = params, cv = kfold, scoring = "neg_mean_squared_log_error")

gcv.fit(dum_X, y)
print(gcv.best_score_)
print(gcv.best_params_)
```

```python
# create dummies

dum_tst = pd.get_dummies(testdf,drop_first = True)

# find y_pred
y_pred = gcv.predict(dum_tst)

res = pd.DataFrame({'id':list(testdf.index), 'Rings': y_pred})
res
```

```python
# make a csv:

res.to_csv("sbt_Abalone_xgb_03.csv", index = False)
```
