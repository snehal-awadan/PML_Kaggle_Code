```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
train = pd.read_csv("train_play.csv",index_col = 0)
train
```

```python
testdf = pd.read_csv("test_play.csv",index_col =0)
testdf
```

```python
le = LabelEncoder()
y = le.fit_transform(train["congestion"])
le.classes_
```

```python
X = train.drop(["congestion"], axis = 1)
X
```

```python
lr = LogisticRegression()
```

```python
kfold = StratifiedKFold(n_splits = 5, shuffle = True, random_state=24)
params = ({"lr_C":np.linspace(0.001,10,10)})

# define gcv:
gcv = GridSearchCV(lr, param_grid = params, cv = kfold, scoring = "neg_mean_squared_log_error")
gcv.fit(X, y)

print(gcv.best_score_)
print(gcv.best_params_)
```
