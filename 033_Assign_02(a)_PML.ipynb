```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("train_Santander.csv",index_col = 0)
testdf = pd.read_csv("test_Santander.csv", index_col = 0)

df.head()
```

```python
y = df["TARGET"]
X = df.drop(['TARGET'], axis = 1)
```

```python
#scaling the data
scaler  = StandardScaler().set_output(transform='pandas')
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
#pca object of PCA()
pca = PCA().set_output(transform='pandas')
p_comps = pca.fit_transform(X_scaled)

pc_cumsum = np.cumsum(pca.explained_variance_ratio_ * 100)
pc_cumsum
```

```python
#a. Finding out how many principal components explain more than 90% variation  
print((len(pc_cumsum)- (pc_cumsum > 90).sum()) + 1 )
```
